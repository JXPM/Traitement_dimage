```python
import cv2
import os
import matplotlib.pyplot as plt 
os.environ["QT_QPA_PLATFORM"] = "xcb"
image = cv2.imread('download.jpeg')
```

```python
plt.imshow(image) 
plt.axis('off')  
plt.show()
```

```python
image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) 
plt.imshow(image_hsv) 
plt.axis('off')   
plt.show()
```

```python
image_cielab = cv2.cvtColor(image, cv2.COLOR_BGR2Lab) 
plt.imshow(image_cielab[:, :, 0], cmap="gray") 
plt.axis('off')   
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 10))
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

axes[0].imshow(image_rgb)
axes[0].set_title("Image RGB")
axes[0].axis("off")

axes[1].imshow(image_cielab)  
axes[1].set_title("CIELAB")
axes[1].axis("off")

axes[2].imshow(image_hsv)
axes[2].set_title("Image HSV")
axes[2].axis("off")

plt.show()
```

```python
img_hsv_mod = image_hsv.copy()
```

```python
img_hsv_mod[:, :, 1] = cv2.subtract(img_hsv_mod[:, :, 1], 90)

img_bgr_mod = cv2.cvtColor(img_hsv_mod, cv2.COLOR_HSV2BGR)

img_rgb_mod = cv2.cvtColor(img_bgr_mod, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb_mod)
plt.axis("off")
plt.title("Image avec saturation diminuée")
plt.show()
```

```python
hue_channel = image_hsv[:, :, 0]
plt.figure(figsize=(18, 6))
plt.subplot(1, 3, 1)
plt.imshow(hue_channel, cmap="gray")
plt.axis("off")
plt.title("Composante Hue (Teinte)")
```

```python
image_hsv_mod = image_hsv.copy()
image_hsv_mod[:, :, 1] = cv2.add(image_hsv_mod[:, :, 1], 50)
```

```python
image_bgr_mod = cv2.cvtColor(image_hsv_mod, cv2.COLOR_HSV2BGR)
image_rgb_mod = cv2.cvtColor(image_bgr_mod, cv2.COLOR_BGR2RGB)
image_rgb_original = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(20, 6))
plt.subplot(1, 3, 2)
plt.imshow(image_rgb_original)
plt.axis("off")
plt.title("Image Originale (RGB)")

plt.subplot(1, 3, 3)
plt.imshow(image_rgb_mod)
plt.axis("off")
plt.title("Image avec saturation augmentée")

plt.show()
```
